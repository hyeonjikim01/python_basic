### 그룹별 집계

```python
import pandas as pd
import numpy as np
```

```python
ipl_data = {'Team' : ['Riders', 'Riders', 'Devils', 'Devils', 'Kings', 'kings',
                      'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
            'Rank': [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
            'Year': [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
            'Points': [876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690]}

df = pd.DataFrame(ipl_data)
df
```

```python
## 팀 별 점수 합계/평균 구하기

df.groupby("Team")["Points"].sum()
```

```python
df.groupby("Team")["Points"].mean()
```

```python
## 다중 인덱스

multi_groupby = df.groupby(["Team", "Year"])["Points"].sum()
multi_groupby
```

```python
## unstack은 행 인덱스를 열로 바꾸는 함수

multi_groupby.unstack()
```

```python
multi_groupby.groupby(level = 0).sum() # leveel = 0은 행 기준
```

```python
multi_groupby.groupby(level = 1).sum() # level = 1은 열 기준
```

```python
# 함수 적용하기_정규화
# lambda x:는 간단한 함수를 정의하는 방식
score = lambda x: (x - x.mean()) / x.std()
df.groupby("Team").transform(score) # 원본 값을 유지하면서 변형
```

```python
## 데이터 필터링

df.groupby("Team").filter(
    lambda x:
    x["Points"].max() > 800) # 최댓값이 800보다 큰 팀을 도출
```

### 데이터 병합하기

```python
# 데이터 준비
raw_data = {
    'subject_id' : ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
    'test_score' : [51, 15, 15, 61, 16,14,15, 1, 61, 16]}

df_left = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_left
```

```python
# 데이터 준비
raw_data = {
    'subject_id': ['4', '5', '6', '7', '8'],
    'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']
}

df_right = pd.DataFrame(raw_data, columns=['subject_id', 'first_name', 'last_name'])
df_right
```

```python
## 교집합

pd.merge(left = df_left, right = df_right, how = "inner", on = 'subject_id') # how = 방식, on = 기준
```

```python
## 한 쪽 데이터 기준으로 병합

# left_join
pd.merge(df_left, df_right, on = 'subject_id', how = 'left')
```

```python
# right_join
pd.merge(df_left, df_right, on = 'subject_id', how = 'right')
```

```python
## 완전 조인

# outer_join, full_join
pd.merge(df_left, df_right, on = 'subject_id', how = 'outer')
```

```python
## 데이터 연결하기

# 데이터 프레임 생성
df1 = pd.DataFrame({
    '이름':['김철수', '이영희', '박민수'], '나이': [25, 30, 28]
})

df2 = pd.DataFrame({
    '이름': ['최지은', '장현우'], '나이': [27, 32]
})

# concat() 세로 방향으로 연결
df_concat = pd.concat([df1, df2], axis = 0, ignore_index = True) # axis = 0은 행 방향, 즉 세로 방향으로 작동

# 결과
print(df_concat)
```
